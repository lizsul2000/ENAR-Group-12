```
import matplotlib.pyplot as plt
import numpy as np

# Data
effect_types = ['Direct Effect', 'Indirect Effect', 'Total Effect']
mediators = ['Alcohol Consumption', 'Smoking','Physical Activity', 'Diabetes', 'Cardiovascular Disease', 'Medication Use*']  # Reversed order

# Coefficients and confidence intervals
coefficients = {
    'Medication Use*': {
        'Direct Effect': 1.065,
        'Indirect Effect': 0.708,
        'Total Effect': 1.773,
        'CI Lower': [0.754, 0.440, 1.335],
        'CI Upper': [1.376, 0.975, 2.210]
    },
    'Cardiovascular Disease': {
        'Direct Effect': 0.955,
        'Indirect Effect': 1.331,
        'Total Effect': 2.287,
        'CI Lower': [0.760, 0.830, 1.745],
        'CI Upper': [1.150, 1.833, 2.828]
    },
    'Diabetes': {
        'Direct Effect': 0.955,
        'Indirect Effect': 1.914,
        'Total Effect': 2.869,
        'CI Lower': [0.760, 1.388, 2.256],
        'CI Upper': [1.150, 2.439, 3.482]
    },
    'Physical Activity': {
        'Direct Effect': 0.955,
        'Indirect Effect': 0.850,
        'Total Effect': 1.805,
        'CI Lower': [0.760, 0.529, 1.457],
        'CI Upper': [1.150, 1.171, 2.153]
    },
    'Smoking': {
        'Direct Effect': 0.955,
        'Indirect Effect': 1.419,
        'Total Effect': 2.375,
        'CI Lower': [0.760, 1.026, 1.907],
        'CI Upper': [1.150, 1.813, 2.842]
    },
    'Alcohol Consumption': {
        'Direct Effect': 0.955,
        'Indirect Effect': 0.916,
        'Total Effect': 1.871,
        'CI Lower': [0.760, 0.669, 1.554],
        'CI Upper': [1.150, 1.163, 2.188]
    }
}

# Colors for effect types
colors = {
    'Direct Effect': 'blue',
    'Indirect Effect': 'green',
    'Total Effect': 'red'
}

# Plot setup
fig, ax = plt.subplots(figsize=(10, 8))

# Y-axis positions for mediators (reversed order)
y_positions = np.arange(len(mediators))

# Plotting
for i, mediator in enumerate(mediators):
    for j, effect_type in enumerate(effect_types):
        coefficient = coefficients[mediator][effect_type]
        ci_lower = coefficients[mediator]['CI Lower'][j]
        ci_upper = coefficients[mediator]['CI Upper'][j]
        
        # Calculate error bar positions
        x = coefficient
        y = y_positions[i] + (j - 1) * 0.2  # Offset for effect types
        
        # Plot error bars
        ax.errorbar(x, y, xerr=[[x - ci_lower], [ci_upper - x]], fmt='o', color=colors[effect_type], capsize=5, label=effect_type if i == 0 else "")

# Customize the plot
ax.set_yticks(y_positions)
ax.set_yticklabels(mediators)
ax.set_xlabel('Odds Ratio')
ax.set_title('Mediation Effect - Odds Ratios with 95% Confidence Intervals')

# Include 0 in the x-axis
ax.set_xlim(left=0)

# Add a vertical line at x = 1
ax.axvline(x=1, color='gray', linestyle='--', linewidth=1, label='Reference (x=1)')

# Add legend
ax.legend(title='Effect Type', loc='upper right')

# Show the plot
plt.tight_layout()
plt.show()
```
